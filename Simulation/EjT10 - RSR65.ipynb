> *Resuleve el problema de la mochila (Knapsack problem) mediante ACO. Este problema, modela una situación análoga al llenar una mochila, incapaz de soportar más de un peso determinado, con todo o parte de un conjunto de objetos, cada uno con un peso y valor específicos. Los objetos colocados en la mochila deben maximizar el valor total sin exceder el peso máximo*.

---

```python
import matplotlib
import pylab
import scipy
import antsys
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from antsys import AntWorld
from antsys import AntSystem
import numpy as np
import random
```

```python
objects = []

for i in range(20):
    weight = random.randint(1, 10)  
    value = random.randint(1, 100)  
    objects.append((i, weight, value))
    print('|%2i|%3i|%4i|' % objects[i])


def knapsack_rules(object, knapsack):
    if knapsack[0] + object[1] <= max_weight:  # Si el objeto cabe en la mochila
        return [(object[0], object[1], object[2])]        
    else:
        return [(0, 0, 0)]



def knapsack_heuristic(path, candidate):
    return candidate.info[1]


def knapsack_cost(path):
    total_value = 0
    for object in path:
        total_value += object.info[2]  # Sumar el valor del objeto
    return total_value


# Mostrar solución
def print_solution(solution):
    print('Total valor en la mochila = %i' % solution[0])
    print('Objetos seleccionados:')
    print('| id | Peso | Valor |')
    for object in solution[2]:
        print('|%3i|%7i|%7i|' % (object.start[0], objects[object.start[0]][1], objects[object.start[0]][2]))


new_world = AntWorld(objects, knapsack_rules, knapsack_cost, knapsack_heuristic)

max_weight = 55

ant_opt = AntSystem(world=new_world, n_ants=50)

ant_opt.optimize(100, 20)

print_solution(ant_opt.g_best)
```
