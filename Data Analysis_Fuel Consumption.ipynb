#### Clases de vehiculos duplicadas

Analizando el dataset, se ha observado como algunas clases de vehiculos están repetidas pero con ligeros cambios en la nomenclatura, por lo que imputan como dos clases completamente diferentes a pesar de ser la misma.

```python
print(sorted(df['VEHICLE CLASS'].unique()))
print(len(df['VEHICLE CLASS'].unique()))
```

Como se observa, hay varias categorias que aparecen repetidas, el primer caso es el que una de las clases aparece escrita completamente en mayusculas mientras que la otra unicamente dispone de la primera letra en mayusculas.

A continuación se va a resolver ese problema pasando todos los registroa a mayuscula.

```python
df['VEHICLE CLASS'] = df['VEHICLE CLASS'].str.upper()
print(sorted(df['VEHICLE CLASS'].unique()))
print(len(df['VEHICLE CLASS'].unique()))
```

Con este ligero cambio a mayusculas se ha reducido en 8 las clases de vehiculos.

Asimismo, se observa como muchas de las clases aparecen escritas con dos puntos y otras de ellas con el mismo significado se escriben con guion, como por ejemplo 'PICKUP TRUCK - SMALL' y 'PICKUP TRUCK: SMALL', a continuación se resolverá estre problema reemplazando los ':' por ' -' utilizando expresiones regulares para que las clases aparezcan escritas de la misma manera.

```python
import re

df['VEHICLE CLASS'] = df['VEHICLE CLASS'].astype(str).apply(lambda x: re.sub(r':\s*', ' - ', x))

print(sorted(df['VEHICLE CLASS'].unique()))
print(len(df['VEHICLE CLASS'].unique()))
```

Con estos cambios aplicados, las clases de vehiculos se reducen de 32 a 17.

#### Creación de la columna 'GEARS'

En apartados anteriores se han explicado los datos presentes en la columna 'TRANSMISSION', es interesante separar esa columna en dos, para analizar tambien numericamente la cantidad de marchas de las que disponen los vehiculos presentes en el dataset.

El numero de marchas viene dado por el ultimo caracter de la celda, por lo que se extraerá el numero mediante expresiones regulares y posteriormente se eliminará de la columna 'TRANSMISSION'.

##### Identificación del parametro **max_depth**

```python
train_scores_r2, test_scores_r2 = [],[]
train_scores_mae, test_scores_mae = [],[]
train_scores_mse, test_scores_mse = [],[]

# Valores del parametro max_depth
values = [i for i in range(1, 30)]

for i in values:
	model = DecisionTreeRegressor(max_depth=i)

	model.fit(x_train, y_train)

	train_predict = model.predict(x_train)
	#R2
	train_r2 = round(r2_score(y_train, train_predict), 3)
	train_scores_r2.append(train_r2)
	#MAE
	train_mae = round(mean_absolute_error(y_train,train_predict),3)
	train_scores_mae.append(train_mae)
	#MSE
	train_mse = round(mean_squared_error(y_train,train_predict),3)
	train_scores_mse.append(train_mse)

	test_predict = model.predict(x_test)
	#R2
	test_r2 = round(r2_score(y_test, test_predict),3)
	test_scores_r2.append(test_r2)
	#MAE
	test_mae = round(mean_absolute_error(y_test, test_predict),3)
	test_scores_mae.append(test_mae)
	#MSE
	test_mse = round(mean_squared_error(y_test,test_predict),3)
	test_scores_mse.append(test_mse)


pyplot.figure(figsize=(18, 6))
# R2
pyplot.subplot(1, 3, 1)
pyplot.plot(values, train_scores_r2, '-o', label='Train (R2)', color='blue')
pyplot.plot(values, test_scores_r2, '-o', label='Test (R2)', color='skyblue')
pyplot.xlabel("Max_depths")
pyplot.ylabel("Scores (R2)")
pyplot.legend()

# MAE
pyplot.subplot(1, 3, 2)
pyplot.plot(values, train_scores_mae, '-o', label='Train (MAE)', color='green')
pyplot.plot(values, test_scores_mae, '-o', label='Test (MAE)', color='lightgreen')
pyplot.xlabel("Max_depths")
pyplot.ylabel("Scores (MAE)")
pyplot.legend()

# MSE
pyplot.subplot(1, 3, 3)
pyplot.plot(values, train_scores_mse, '-o', label='Train (MSE)', color='red')
pyplot.plot(values, test_scores_mse, '-o', label='Test (MSE)', color='salmon')
pyplot.xlabel("Max_depths")
pyplot.ylabel("Scores (MSE)")
pyplot.legend()

pyplot.tight_layout()
pyplot.show()
```

```python
n_clusters = 5

clusterer = Pipeline(
    [
        (
            "kmeans",
            KMeans(
                n_clusters=n_clusters,
                init="k-means++",
                n_init=50,
                max_iter=500,
                random_state=9,
            ),
        ),
    ]
)
```

A continuación se define la pipeline y se le ajustan los datos.

```python
pipe = Pipeline([("preprocessor", preprocessor), ("clusterer", clusterer)])
pipe.fit(X)
```

```python
preprocessed_data = pipe["preprocessor"].transform(X)
preprocessed_data
```

```python
predicted_labels = pipe["clusterer"]["kmeans"].labels_
predicted_labels
```

En este punto los datos del conjunto X están preprocesados y el clusterer ha definido una serie de etiquetas para la agrupación de las instancias.

A continuación se imprimirán por pantalla una serie de metricas que ayudarán a comprender como de bien está funcionando el clusterer.

```python
print(silhouette_score(preprocessed_data, predicted_labels))
print(adjusted_rand_score(y, predicted_labels))
```

El Silhouette Score de aproximadamente 0.51 indica una separación moderada entre los clusters, mientras que el Adjusted Rand Score de aproximadamente 0.08 indica una similitud muy baja entre los agrupamientos predichos y los verdaderos, lo que indica que los clusters encontrados por el algoritmo de clustering no coinciden bien con las etiquetas verdaderas.

Finalmente, se va a ejecutar nuevamente el algoritmo de clustering con el objetivo de almacenar los resultados en un dataframe para la visualización de las agrupaciones.

```python
pcadf = pd.DataFrame(
    pipe["preprocessor"].transform(X),
    columns=["component_1", "component_2"],
)

pcadf["predicted_cluster"] = pipe["clusterer"]["kmeans"].labels_
pcadf["true_label"] = y

plt.style.use("fivethirtyeight")
plt.figure(figsize=(8, 8))

scat = sns.scatterplot(
    x="component_1",
    y="component_2",
    s=50,
    data=pcadf,
    hue="predicted_cluster",
    style="true_label",
    palette="Set2",
)

scat.set_title("Resultados del pipeline de Clustering")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.0)

plt.show()
```

### Hierarchical Clustering

A continuación y para finaliar el apartado de clustering se va a aplicar la técnica de clustering jerárquico, un método de agrupamiento de datos que busca construir una jerarquía de grupos. Para la agrupación de las diferentes instancias, se utiliza la distancia euclidea para determinar la similitud o diferencia entre los puntos de datos. Cuanto más pequeña sea la distancia euclidea entre dos puntos, más similares se consideran en términos de su ubicación en el espacio y mas probable será añadirlos a un mismo cluster.

La principal ventaja de este método es el la de no tener que especificar previamente el número de clusters, lo que lo hace útil cuando no se tiene información sobre la estructura de los datos y el resultado de esta estrategia es un grafico en forma de arbol llamado dendrograma. 

También, al ser una estrategia computacionalmente costosa, se va a aplicar sobre un conjunto de caracteristicas reducido.

En este caso se ha decidido aplicar el clustering jerarquico únicamente sobre las columnas Marca, modelo, Pais y Marchas, esta última añadida para disponer también de una caracteristica numerica.

```python
df_hier = df_clus.loc[:, ['MAKE', 'MODEL', 'COUNTRY', 'GEARS']]

scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_hier)

Z = linkage(df_scaled, method='ward')

plt.figure(figsize=(12, 6))
plt.title('Dendrograma Jerárquico')
plt.xlabel('Índice de Muestra')
plt.ylabel('Distancia Euclidiana')
dendrogram(
    Z,
    leaf_rotation=90.,  
    leaf_font_size=8., 
)
plt.show()
```

En el grafico anterior se muestra el dendrograma creado.

La interpretación del grafico se hace desde la parte superior a la inferior, como se observa en la primera división (color azul) se generan tres clusters que contienen la totalidad de las instancias.

Posteriormente cada uno de estos clusters realizan sus propias divisiones para crear subgrupos con instancias mas parecidas, basandose en la distancia euclidea que presentan entre ellas basandose en la distancia euclidea.
